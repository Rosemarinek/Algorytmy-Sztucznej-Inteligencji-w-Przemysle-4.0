```python
import gym
from stable_baselines3 import DQN


model = DQN("MlpPolicy", "LunarLander-v2", verbose=1, exploration_final_eps=0.1, target_update_interval=250)
model.learn(total_timesteps=int(1e5), log_interval=10)
model.save("model")
```

```python
import gnwrapper

env = gnwrapper.Monitor(gym.make("LunarLander-v2", render_mode='rgb_array'),directory="./video/")


observation = env.reset()
total_reward = 0
for i in range(600):
    action, _states = model.predict(observation)
    observation, reward, terminated, truncated = env.step(action)
    total_reward += reward


env.display()
print(f"Total reward: {total_reward}")
```
